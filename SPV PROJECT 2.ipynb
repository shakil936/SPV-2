```python
df= pd.read_csv('/content/drive/MyDrive/svp5/WorldCupMatches.csv')
df
```

```python
df["Home Team Goals"].fillna(df["Home Team Goals"].mean(),inplace=True)
df
```

```python
df.dropna(inplace=True)
df
```

```python
df.to_csv('/content/drive/MyDrive/svp5/worldCup.csv',index=True)
```

```python
cup_column=pd.Series([9,9,9,9,9,9,9])
df["Cup"]=cup_column
df
```

```python
df["Cup"].fillna(9,inplace=True)
df.head(10)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

x = list(df['Home Team Name'])
y = list(df['Home Team Goals'])

x_pos = [i for i, _ in enumerate(x)]

plt.bar(x_pos, y, color='red')
plt.xlabel("Home Team Name")
plt.ylabel("Home Team Goals")
plt.title("Home Team Name by Home team Goals")

plt.xticks(x_pos, x)

plt.show()
```
